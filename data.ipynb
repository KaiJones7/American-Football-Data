```python
from scipy import stats

# Perform linear regression
x = df['year']
y = df['pass yards']
slope, intercept, r_value, p_value, std_err = stats.linregress(x , y)
if p_value < 0.05:
    print("The trend in pass yards is significant.")
else:
    print("The trend in pass yards is not significant.")

plt.scatter(x, y)

plt.plot(slope, color='red', label='Regression Line')
plt.show()
```
